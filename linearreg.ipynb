```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load features
X = pd.read_csv("final_merged_dataset_.csv")
X["Hour"] = pd.to_datetime(X["Hour"])

# Sinusoidal time features
X["hour_sin"] = np.sin(2 * np.pi * X["Hour"].dt.hour / 24)
X["hour_cos"] = np.cos(2 * np.pi * X["Hour"].dt.hour / 24)
X["dayofyear_sin"] = np.sin(2 * np.pi * X["Hour"].dt.dayofyear / 365)
X["dayofyear_cos"] = np.cos(2 * np.pi * X["Hour"].dt.dayofyear / 365)

# Loop for each label (rate class)
rate_classes = ["CIEP", "GPC", "GPI", "GSC", "GSI", "GSTC", "GSTI", "GTC", "GTI", "Res", "RSCP"]
for rate in rate_classes:
    print(f"\nLinear Regression for {rate}")
    
    # Load label
    y = pd.read_csv(f"{rate}_cleaned.csv")
    y["Date"] = pd.to_datetime(y["Date"])
    
    # Merge features and label
    df = pd.merge(X, y, left_on="Hour", right_on="Date", how="inner")
    
    # Drop customerAcc_number_monthly
    df = df.drop(columns=["Hour", "Date"], errors="ignore")
    if "customerAcc_number_monthly" in df.columns:
        df = df.drop(columns=["customerAcc_number_monthly"])
    
    # Find label column
    label_col = [col for col in df.columns if col not in X.columns][0]
    X_feature = df.drop(columns=[label_col])
    y_label = df[label_col]
    
    # Split, fit, predict
    X_train, X_test, y_train, y_test = train_test_split(X_feature, y_label, test_size=0.2, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Print metrics
    print(f"RÂ²: {r2_score(y_test, y_pred):.4f}")
    print(f"MSE: {mean_squared_error(y_test, y_pred):.4f}")
    
    # Print coefficients
    coeff_df = pd.DataFrame({
        "Feature": X_feature.columns,
        "Coefficient": model.coef_
    })
    print("Coefficients:")
    print(coeff_df)

   # Plot Actual vs Predicted
    plt.figure(figsize=(8,6))
    sns.scatterplot(x=y_test, y=y_pred)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "r")
    plt.xlabel(f"Actual {rate}")
    plt.ylabel(f"Predicted {rate}")
    plt.title(f"Linear Regression: Predicted vs Actual ({rate})")
    plt.show()
```
