```python
import pandas as pd
import pyxlsb
```

```python
def merge_csv_and_excel(csv_path, excel_path):
    csv = pd.read_csv(
        csv_path, 
        header=4,
        names=['date'] + [f"hour{i}" for i in range(1, 26)] # Rename column headers to: date, hour1,..., hour25
    )

    excel = pd.read_excel(
        excel_path,
        header=5,
        names=['date'] + [f"hour{i}" for i in range(1, 26)] # Rename column headers to: date, hour1,..., hour25
    )

    merged = pd.concat([csv, excel], ignore_index=True)
    merged['date'] = pd.to_datetime(merged['date'])

    return merged

CIEP = merge_csv_and_excel(
    csv_path='Data/Predict - y/JCPL/CIEP/Hist_JCP&L_CIEP-Eligible_Hourly_Load.csv', 
    excel_path='Data/Predict - y/JCPL/CIEP/JCP&L_CIEP-Eligible_Hourly_Load_August_2025_Update.xlsx'
)

RSCP = merge_csv_and_excel(
    csv_path='Data/Predict - y/JCPL/RSCP/Hist_JCP&L_RSCP-Eligible_Hourly_Load.csv',
    excel_path='Data/Predict - y/JCPL/RSCP/JCP&L_RSCP-Eligible_Hourly_Load_August_2025_Update.xlsx'
)
```
